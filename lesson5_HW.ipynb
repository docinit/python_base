# Задача 1

```python
print('''В этой программе пользователь вводит свой текст, который сохраняется в файл.
Если пользователь нажимает клавишу Enter без ввода текста, программа завершает свою работу.
''')
string = input('Введите текст для сохранения: ')
while string:
    file = open('lesson5_HW.md','a')
    file.writelines(string+'\n')
    file.close()
    string = input('Введите текст для сохранения: ')
```

```python
# Чтение сохраненного файла
with open('lesson5_HW.md','r') as f:
    print(*[i.replace('\n','') for i in f.readlines()],sep='\n')
```

# Задача 3

```python
print('''В этой программе в файл сохраняется заранее подготовленный список фамилий сотрудников и суммы дохода;
в каждой строке - одна фамилия и доход человека; значения одной строки разделены запятыми;
Далее производится чтение файла и выводится список фамилий людей, у которых доход ниже заданного порога.
В конце программы выводится среднее значение оклада всех сотрудников.
''')

salaries = \
'''Иванов,20000
Петров,15500
Сидоров,25000
Алексеев,30000
Стрижов,17000
Соколов,25000
Орлов,22000
Ермолаев,18500
Емельянов,19000
Семенов,65000'''

with open('lesson5_HW_task3','w') as f:
    f.writelines(salaries)
    
with open('lesson5_HW_task3','r') as f:
    salary_list = [[i.split(',')[0],\
                    int(i.split(',')[1].replace('\n',''))]\
                   for i in f.readlines()]
    
threshold = 20000
print(f'Следующие сотрудники имеют оклад менее {threshold} рублей:')
for k,i in enumerate(salary_list):
    if i[1]<threshold:
        print(i[0])
        
print(f'Средний оклад сотрудников составил {sum([i[1] for i in salary_list])/len(salary_list):.2f}')
```

# Задача 7

```python
print('''В этой программе пользователь вводит данные о названии, форме собственности, сумме доходов и расходов фирмы;
процесс ввода новых данных прекращется при введении любого знака в специальном поле;
Введенные пользователем данные сохраняются в файл; после этого производится
подсчет суммы дохода с учетом расходов (прибыли);
список компаний с их прибылью, а также средний уровень дохода
компаний с положительной прибылью,
сохраняются в словарь.
''')

do_not_proceed = None
while not do_not_proceed:
    firm = input('Введите название фирмы: ')
    prop_type = input('Укажите форму собственности: ')
    incomes = input('Укажите сумму доходов: ')
    expences = input('Укажите сумму расходов: ')
    with open('lesson5_HW_task7.txt','a') as f:
        print(firm,prop_type,incomes,expences,file=f,sep=',')
    do_not_proceed = input('Введите любой знак для завершения работы программы: ')
    if do_not_proceed == '':
        do_not_proceed == None
```

```python
income_dict = {}
total_profit = 0
i=0
with open('lesson5_HW_task7.txt','r') as f:
    for line in f:
        firm_info = line.split(',')
        profit = float(firm_info[-2]) - float(firm_info[-1])
        print(f'Доход компании {firm_info[0]} = {profit}')
        income_dict[firm_info[0]] = profit
        if profit >=0:
            total_profit += profit
            i+=1
    income_avg = {'average_profit': total_profit / i}
result_list = [income_dict,income_avg]
result_list
```

```python
import json
with open('lesson5_HW_task7_result_list.json','w') as f:
    json.dump(result_list,f,ensure_ascii=False)
```
